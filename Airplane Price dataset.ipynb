```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("airplane_price_dataset.csv")
df
```

```python
sns.set(style="whitegrid")
numerical_cols = ["Üretim Yılı","Motor Sayısı",	"Kapasite",	"Menzil (km)", "Yakıt Tüketimi (L/saat)", "Saatlik Bakım Maliyeti ($)", "Yaş", "Fiyat ($)"]
categorical_cols = ["Model","Motor Türü","Satış Bölgesi"]
```

# Plot histograms for each numerical feature

```python
df.hist(bins=30, figsize=(15, 10))
plt.show()
```

# This is subplot has show Distribution of Numerical Features

```python

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
fig.suptitle("Distribution of Numerical Features", fontsize=16)
axes = axes.flatten()

# Plot histograms
for i, col in enumerate(numerical_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i])  # Correct axis indexing
    axes[i].set_title(col)  # Correct title setting

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Correct rect syntax
plt.show()
```

# This plot has show the Larger airplanes with higher Passenger capacity tend

```python
plt.figure(figsize=(15,6))
ax=sns.barplot(x="Model",y="Kapasite",data=df, hue="Motor Türü")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

# The percantage of fuel/Model and The percantage of KM/Model

```python
# Aggregate both fuel consumption and range per motor type
df_grouped = df.groupby("Model")[["Yakıt Tüketimi (L/saat)", "Menzil (km)"]].sum()

# Plotting pie charts side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for fuel consumption
axes[0].pie(df_grouped["Yakıt Tüketimi (L/saat)"], labels=df_grouped.index, autopct="%0.1f%%", radius=1.2, labeldistance=1.1)
axes[0].set_title("Model Bazında Yakıt Tüketimi Dağılımı")

# Pie chart for range
axes[1].pie(df_grouped["Menzil (km)"], labels=df_grouped.index, autopct="%0.1f%%", radius=1.2, labeldistance=1.1)
axes[1].set_title("Model Bazında Menzil (km)")

plt.tight_layout()
plt.show()
```

# Count Plots for Categorical Features

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Count Plots for Categorical Features", fontsize=16)

for i, col in enumerate(categorical_cols):
    sns.countplot(y=df[col], order=df[col].value_counts().index[:10], ax=axes[i])  # Show top 10 categories
    axes[i].set_title(f"Top 10 {col}")

plt.tight_layout()
plt.show()
```
